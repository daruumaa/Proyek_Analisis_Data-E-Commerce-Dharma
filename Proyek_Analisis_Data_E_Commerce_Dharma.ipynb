## Visualization & Explanatory Analysis

### Pertanyaan 1 : Bagaimana distribusi lokasi geografis pelanggan dalam dataset?

Mengidentifikasi Kode ZIP dengan Banyak Negara Bagian

```python
zip_code_state_counts = geo_df.groupby('geolocation_zip_code_prefix')['geolocation_state'].nunique().reset_index(name='state_count')
zip_codes_with_multiple_states = zip_code_state_counts[zip_code_state_counts['state_count'] >= 2].shape
```

Menentukan Negara Bagian Paling Sering untuk Setiap Kode ZIP

```python
most_frequent_state = geo_df.groupby(['geolocation_zip_code_prefix', 'geolocation_state']).size().reset_index(name='state_frequency')
most_frequent_state = most_frequent_state.sort_values(by='state_frequency', ascending=False).drop_duplicates(subset='geolocation_zip_code_prefix').drop('state_frequency', axis=1)
```

Menghitung Median Geolokasi untuk Setiap Kode ZIP, Kota, dan Negara Bagian

```python
median_geolocation = geo_df.groupby(['geolocation_zip_code_prefix', 'geolocation_city', 'geolocation_state'])[['geolocation_lat', 'geolocation_lng']].median().reset_index()
```

Menyaring Berdasarkan Negara Bagian Paling Sering untuk Setiap Kode ZIP

```python
filtered_geolocation = median_geolocation.merge(most_frequent_state, on=['geolocation_zip_code_prefix', 'geolocation_state'], how='inner')
```

Menggabungkan dengan Data Pelanggan

```python
customers_silver = customers_df.merge(filtered_geolocation, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='inner')
```

```python
customers_silver.head()
```

```python
def plot_peta_brasil(data):
    # Memuat peta Brasil
    url_peta_brasil = 'https://i.pinimg.com/originals/3a/0c/e1/3a0ce18b3c842748c255bc0aa445ad41.jpg'
    peta_brasil = mpimg.imread(urllib.request.urlopen(url_peta_brasil), 'jpg')

    # Membuat plot scatter
    ax = data.plot(kind="scatter", x="geolocation_lng", y="geolocation_lat", figsize=(10,10), alpha=0.3, s=0.3, c='blue')

    # Menghilangkan sumbu
    plt.axis('off')

    # Menampilkan peta Brasil
    plt.imshow(peta_brasil, extent=[-73.98283055, -33.8, -33.75116944, 5.4])

    # Menampilkan plot
    plt.show()
```

```python
# Memplot peta menggunakan data pelanggan yang telah difilter
plot_peta_brasil(customers_silver.drop_duplicates(subset='customer_unique_id'))
```

### Pertanyaan 2 : roduk kategori apa yang paling banyak terjual dan menghasilkan pendapatan tertinggi?

```python
order_items = pd.read_csv('/content/order_items_dataset.csv')
products = pd.read_csv('/content/products_dataset.csv')
sellers = pd.read_csv('/content/sellers_dataset.csv')
product_category_translation = pd.read_csv('/content/product_category_name_translation.csv')
```

```python
# Menggabungkan dataset produk dengan terjemahan kategori produk
products = products.merge(product_category_translation, on='product_category_name', how='left')

# Menggabungkan dataset order_items dengan products
order_product = order_items.merge(products, on='product_id', how='left')

# Menggabungkan dataset order_product dengan sellers
order_product_seller = order_product.merge(sellers, on='seller_id', how='left')
```

```python
# Menghitung jumlah dan total pendapatan per kategori produk
category_sales = order_product.groupby('product_category_name_english').agg({
    'order_item_id': 'count',
    'price': 'sum'
}).reset_index()

category_sales = category_sales.rename(columns={'order_item_id': 'total_sales', 'price': 'total_revenue'})
```

```python
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
category_sales.sort_values(by='total_sales', ascending=False).head(10).plot(
    x='product_category_name_english', y='total_sales', kind='bar', ax=plt.gca(), color='skyblue')
plt.title('Top 10 Kategori Produk Berdasarkan Jumlah Penjualan')
plt.ylabel('Jumlah Penjualan')
plt.xlabel('Kategori Produk')

plt.subplot(1, 2, 2)
category_sales.sort_values(by='total_revenue', ascending=False).head(10).plot(
    x='product_category_name_english', y='total_revenue', kind='bar', ax=plt.gca(), color='lightgreen')
plt.title('Top 10 Kategori Produk Berdasarkan Pendapatan')
plt.ylabel('Total Pendapatan')
plt.xlabel('Kategori Produk')

plt.tight_layout()
plt.show()
```

### Pertanyaan 3 : Bagaimana distribusi penjualan berdasarkan lokasi penjual, dan apakah ada hubungan antara lokasi penjual dengan jumlah produk yang terjual?

```python
# Menghitung jumlah penjualan per penjual berdasarkan lokasi
seller_sales_location = order_product_seller.groupby('seller_city').agg({
    'order_item_id': 'count',
    'price': 'sum'
}).reset_index()

seller_sales_location = seller_sales_location.rename(columns={'order_item_id': 'total_sales', 'price': 'total_revenue'})
```

```python
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
seller_sales_location.sort_values(by='total_sales', ascending=False).head(10).plot(
    x='seller_city', y='total_sales', kind='bar', ax=plt.gca(), color='orange')
plt.title('Top 10 Kota Berdasarkan Jumlah Penjualan')
plt.ylabel('Jumlah Penjualan')
plt.xlabel('Kota Penjual')

plt.subplot(1, 2, 2)
seller_sales_location.sort_values(by='total_revenue', ascending=False).head(10).plot(
    x='seller_city', y='total_revenue', kind='bar', ax=plt.gca(), color='purple')
plt.title('Top 10 Kota Berdasarkan Pendapatan')
plt.ylabel('Total Pendapatan')
plt.xlabel('Kota Penjual')

plt.tight_layout()
plt.show()
```
